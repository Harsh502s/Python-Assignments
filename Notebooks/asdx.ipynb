```python
from scipy.cluster.hierarchy import dendrogram, linkage
df = pd.read_csv(r"C:\Users\ds12\Downloads\mtcars.csv",index_col=0)
df.head()
```

```python
z = linkage(df, 'ward')

dendrogram(z, labels=df.index, leaf_rotation=90, leaf_font_size=8,color_threshold=200);

# Adding the color threshold line
plt.axhline(y=200, color='r', linestyle='--')
```

```python
sns.clustermap(df, method='single', metric='correlation', cmap='viridis')
```
