# Dataset-Heart Disease

### Load csv and set value -9 to NA

```python
heart=pd.read_csv(r"C:\Users\harsh\College\Python-Assignments\Dataset\heartdiseases.csv")
heart.head()
```

```python
heart=heart.replace(-9,np.nan)
heart.head()
```

### Processing will be focused on first 58 columns.

a. Take subset of first 58 columns

```python
heart=heart.iloc[:,0:58]
heart
```

### Values 58 column defines how many times a patient had a heart attack

a. Since for building predictive model, let’s focus on only 2 values: 0 means no attack, 1: possibility of attack. Therefore values > 1 should be reset to 1

```python
heart['V58']=np.where(heart['V58']>1,1,0)
heart
```

### Since its a binary predictive model, we will process only numeric data

a. Check data types of each column

b. Drop columns which are having strings or objects

#### a Check data types of each column

```python
heart.dtypes.value_counts()
```

#### b Drop columns which are having strings or objects

```python
heart2=heart.select_dtypes(include=np.number)
heart2.dtypes
```

### Drop columns which have all null values

a. Check columns which have all null values // is.na, all()

V5', 'V6', 'V45', 'V46', 'V47', 'V48', 'V49', 'V52', 'V53

b. Drop the columns which have all null data

#### a Check columns which have all null values // is.na, all()

```python
heart.isna().all().value_counts()
```

#### b Drop the columns which have all null data

```python
heart=heart.dropna(axis=1,how='all')
heart.head()
```

```python
heart.isnull().all().value_counts()
```

### Find rows with missing values greater than 50%

```python
print('rows with missing values greater than 50%:\n')
for i in range(len(heart.index)) :
    if heart.iloc[i].isnull().sum()/len(heart.columns)>0.5:
        print("Row", i )
```

```python
# Finding rows with missing values greater than 50%
print('rows with missing values greater than 50%:\n')
for i in range(len(heart.index)) :
    if heart.iloc[i].isnull().sum()/len(heart.columns)>0.5:
        print("Row", i )
```

### Fill missing values with mode value for all columns which are float or int

a. Print columns that have minimum one NaN value

b. For each value, Update Nan values with median value of that column

Note: character vectors and factors are imputed with the mode. Numeric and integer vectors are imputed with the median.

c. Verify that no more Na values

#### a Print columns that have minimum one NaN value

```python
null_1=heart.isna().any()
```

```python
print(null_1[null_1].index)
```

#### b For each value, Update Nan values with mode value of that column

```python
heart=heart.fillna(heart.median())
```

```python
heart
```

#### c Verify that no more Na values

```python
heart.isna().sum().sort_values(ascending=False)
```

### Remove columns which have same value in all rows // since such column can not contribute to prediction

```python
nunique = heart.nunique()
cols_to_drop = nunique[nunique == 1].index
heart=heart.drop(cols_to_drop, axis=1)
heart.head()
```
