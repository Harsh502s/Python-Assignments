## Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with  Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelor.

```python
married_count1 = (
    df50kup["marital-status"]
    .map(
        lambda x: x
        in ["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"]
    )
    .value_counts()[True]
)
married_count2 = (
    df50kup["marital-status"]
    .map(
        lambda x: x
        in ["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"]
    )
    .value_counts()[False]
)

print(married_count1)
print(married_count2)
```

```python
print(
    "Proportion of married men who earn more than 50k is ",
    (married_count1 / (married_count1 + married_count2)) * 100,
    "percent",
)
```

```python
print(
    "Proportion of single men who earn more than 50k is ",
    (married_count2 / (married_count1 + married_count2)) * 100,
    "percent",
)
```

# Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan?

### Min hours per week of each native country

```python
df[["native-country", "hours-per-week"]].groupby("native-country").min()
```

### Maximum hour per week for each country.

```python
df[["native-country", "hours-per-week"]].groupby("native-country").max()
```
