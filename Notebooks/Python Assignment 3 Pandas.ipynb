### We'd like to host these wine reviews on our website, but a rating system ranging from 80 to 100 points is too hard to understand - we'd like to translate them into simple star ratings.
### A score of 95 or higher counts as 3 stars, a score of at least 85 but less than 95 is 2 stars. Any other score is 1 star.
### Also, the Canadian Vintners Association bought a lot of ads on the site, so any wines from Canada should automatically get 3 stars, regardless of points.
### Create a series star_ratings with the number of stars corresponding to each review in the dataset

```python
# Create a series with the frequency of "tropical" and "fruity" in wine descriptions
tropical_count = (
    reviews["description"].map(lambda desc: "tropical" in desc.lower()).value_counts()
)
fruity_count = (
    reviews["description"].map(lambda desc: "fruity" in desc.lower()).value_counts()
)

# Print the results
print("Number of descriptions containing 'tropical':", tropical_count[True])
print("Number of descriptions containing 'fruity':", fruity_count[True])
```

```python
troporfruit = {}
fruitcount = 0
tropcount = 0
for i in reviews["description"]:
    if "tropical" in i.lower():
        tropcount += 1
        troporfruit["tropical"] = fruitcount
        pass
    elif "fruity" in i.lower():
        fruitcount += 1
        troporfruit["fruity"] = tropcount
```

```python
troporfruit
```

```python
pd.Series(troporfruit)
```

```python
def compute_rating(score, country):
    if country == "Canada":
        return 3
    elif score >= 95:
        return 3
    elif score >= 85:
        return 2
    else:
        return 1


# apply the function to create the star_ratings column
reviews["star_rating"] = reviews.apply(
    lambda x: compute_rating(x["points"], x["country"]), axis=1
)
```
