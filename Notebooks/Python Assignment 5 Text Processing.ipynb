# Importing the libraries

```python
import pandas as pd
import numpy as np
import re
import string
```

```python
# getting the text for extraction

text = """From: keith@cco.caltech.edu (Keith Allan Schneider)
Subject: Re: <<Pompous ass\nOrganization: California Institute of Technology, Pasadena\nLines: 16\nNNTP-Posting-Host: punisher.caltech.edu\nlivesey@solntze.wpd.sgi.com (Jon Livesey) writes:\n[...]\n>>The "`little' things" above were in reference to\nGermany, clearly. People\n>>said that there were similar things\nin Germany, but no one could name any.\n>That's not true. I\ngave you two examples. One was the rather\n>pevasive anti\nsemitism in German Christianity well before Hitler\n>arrived. \nThe other was the system of social ranks that were used\n>in\nImperail Germany and Austria to distinguish Jews from the\nrest \n>of the population.\nThese don't seem like "little\nthings" to me. At least, they are orders\nworse than the\nmotto. Do you think that the motto is a "little thing"\nthat\nwill lead to worse things?\nkeith
"""
text
```

```python
len(text)
```

```python
# Removing the punctuations that separates the sentences

text1 = text.translate(str.maketrans('', '', string.punctuation))
text1
```

```python
# Extracting the name and email address from the text

name = re.findall(r'From: (.*)\(', text)
print(name)

email = re.findall(r'\((.*)\)', text)
print(email)
```

```python
# Splitting the text by delimiters \n,\t,whitespaces and then lowercasing the text

text2 = text1.split()
text2

text3 = [x.lower() for x in text2]
text3
```

```python
# Removing token which does not contain alphanumeric characters

text4 = [x for x in text3 if x.isalnum()]
text4
```

```python
# Cleaning leading and trailing punctuations

text5 = [x.strip(string.punctuation) for x in text4]
text5
```

```python
import nltk
nltk.download('stopwords')
```

```python
# Import stopwords from wordcloud library

from wordcloud import STOPWORDS

# Removing stopwords from the text

text6 = [x for x in text5 if x not in STOPWORDS]
text6
```

```python
# Removing the stop words

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
text7 = [x for x in text5 if x not in stop_words]
text7
```

```python
# Converting tokens into text again

text8 = ' '.join(text6)
print(text8)
print('\n')
text9 = ' '.join(text7)
print(text9)
```
