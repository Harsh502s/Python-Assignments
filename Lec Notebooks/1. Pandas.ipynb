```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=["a", "b", "c", "d"])
print(data)

data["b"]
```

## Series as specialized dictionary

```python
population_dict = {
    "California": 38332521,
    "Texas": 26448193,
    "New York": 19651127,
    "Florida": 19552860,
    "Illinois": 12882135,
}
population = pd.Series(population_dict)
population
```

# Constructing Data Frame


```python
area_dict = {
    "California": 423967,
    "Texas": 695662,
    "New York": 141297,
    "Florida": 170312,
    "Illinois": 149995,
    "Colorado": 999999,
}
area = pd.Series(area_dict)
area
```

```python
states = pd.DataFrame({"population": population, "area": area})
states
```

```python
print(states.index)
print(states.columns)
print(states.shape)
```

#Index Object

```python
ind = pd.Index([2, 3, 5, 7, 11])
print(ind)
print(type(ind))
print(ind.shape)
```

```python
print(ind[1])
print(ind[::2])
print(ind.size, ind.shape, ind.ndim, ind.dtype)
```

**Index are immutable**

```python
ind[1] = 0
```

**Index as ordered set**

```python
indA = pd.Index([1, 3, 5, 7, 9])
indB = pd.Index([2, 3, 5, 7, 11])

inter = indA & indB  # intersection
print(inter)
un = indA | indB  # union
print(un)
dif = indA ^ indB  # symmetric difference
print(dif)
```
