```python
# Read html file for text extraction
from bs4 import BeautifulSoup
import requests
import re
import pandas, numpy

with open('data.html', 'r') as f:
    contents = f.read() 
    soup = BeautifulSoup(contents, 'html.parser')
    html_data = soup.find_all('div', class_='nb__7nqQI')
    
```

```python
scraped_data = {}
links = []
minimumRent = []
minimumDeposit = []
builtup_area = []
furnishing = []
apartment_type = []
for i in html_data:
    minimumRent.append(i.find("div", attrs={"class": "flex", "id": "minimumRent"}).text)
    minimumDeposit.append(
        i.find(
            "div", attrs={"class": "font-semi-bold heading-6", "id": "roomType"}
        ).text
    )
    builtup_area.append(
        i.find(
            "div", attrs={"class": "flex", "id": "unitCode"}
        ).text
    )
    furnishing.append(
        i.find("div", attrs={"class": "font-semibold"}).text
    )
    apartment_type.append(i.find('div',attrs={"class": "flex flex-1 pl-0.5p"}).find("div", attrs={"class": "font-semibold"}).text
    )
    
    
```
