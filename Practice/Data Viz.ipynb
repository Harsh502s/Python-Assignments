```python
data = countries.groupby('country')['population'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,10))
plt.bar(data.index,data.values)
plt.tick_params(axis='x',rotation=90)
plt.title('Top 10 countries with highest population')
plt.xlabel('Countries')
plt.ylabel('Population in Millions')
# Adding the values on top of the bars
for i in range(len(data)):
    plt.text(x=i,y=data.values[i],s=round(data.values[i]/100000,2),ha = 'center',va='bottom',color='black',fontsize=10)
plt.show()
```

```python
plt.stem(data.index,data.values)
plt.tick_params(axis='x',rotation=90)
```

```python
all_countries = countries.groupby('country')['population'].sum().sort_values(ascending=False)
plt.stem(all_countries.values,all_countries.index)
```
